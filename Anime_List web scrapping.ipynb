### Import necessary libraries

```python
from bs4 import BeautifulSoup as bs
import requests
```

#### Load webpage

```python
#Loading the webpage with anime list using 'request' library 
r=requests.get("https://myanimelist.net/topanime.php")
#reading the content with bautiful soup library
soup=bs(r.content)
contents=soup.prettify()
print(contents)
```

```python
#find the table with anime titles in html sript and narrowing down elements to scrape
table=soup.find(class_="top-ranking-table")
title_rows=table.find_all("h3")
for row in title_rows:
    print(row.prettify()) #printed all elements to scrape 
```

```python
#After exploring the available information, storing all titles and the web page links for the titles in first page.
anime_list=[]
for row in title_rows:
    anime_link=[]
    anime_link.append(row.find("a").get_text())
    anime_link.append(row.a['href'])
    anime_list.append(anime_link)
anime_list #Retriving title and respective web page link for one page succesful
```

#### Following the above example scrapping all the titles and links 

```python
from IPython.display import clear_output #to print current progress to of the for loop
```

```python
anime_list1=[]
#Each webpage has 50 titles and total 18349 titles were present running the loop to scrape info of all anime titles
for i in range(0,18301,50): 
    r1=requests.get("https://myanimelist.net/topanime.php?limit="+str(i))
    table1=(bs(r1.content)).find_all("h3")
    clear_output(wait=True)
    print("current progress:",i)
    for row in table1:
        anime_link=[]
        anime_link.append(row.find("a").get_text())
        anime_link.append(row.a['href'])
        anime_list1.append(anime_link)
#Successfully scraped all the titles and the web links to scrape info for each anime
```

```python
anime_list1 #one unwanted element 'more' was included as title, which repeated multiple times
```

```python
len(anime_list1)
mylist=[]
mylist=anime_list1
```

```python
len(mylist)
```

```python
#Deleting the unwanted item 'More'from the list (cleaning the scrapped data)
for i in range(0 , len(anime_list1)):
    if anime_list1[i][0]=="More":
        anime_list1.pop(i)   
```
